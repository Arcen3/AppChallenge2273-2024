```python
def listsplit(a):
    if len(a) < 2:
        return a
    mid = len(a) // 2
    left = a[:mid]
    right = a[mid:]
    return merge_sort(left=listsplit(left), right=listsplit(right))

def merge_sort(left, right):
    if len(left) == 0:
        return right
    elif len(right) == 0:
        return left
    
    result = []
    ir, il = 0, 0
    
    while ir < len(right) and il < len(left):
        if left[il][4] <= right[ir][4]:
            result.append(left[il])
            il += 1
        else:
            result.append(right[ir])
            ir += 1
            
    # Add remaining elements
    result.extend(left[il:])
    result.extend(right[ir:])
    
    return result

# Example paths list
paths = [
    "path_1_foo_bar_3_2",
    "path_2_baz_qux_4_1",
    "path_3_hello_world_5_3",
    "path_4_test_case_6_2"
]

indexed_path = [i.split('_') for i in paths]
sortpaths = listsplit(indexed_path)
print(sortpaths)
```
